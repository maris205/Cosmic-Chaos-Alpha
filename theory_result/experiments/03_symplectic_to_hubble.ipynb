# Notebook 3: Symplectic Volume → Hubble Evolution

**Paper reference**: §II.H, Proposition (Symplectic derivation of the Hubble law)

**Goal**: Verify numerically that $(1/\varepsilon)(d\varepsilon/dn) = -2/(n\ln n)$ and that this gives the Hubble evolution $H(t) = H_\infty + \beta/\ln^2(t/t_P)$.

```python
import numpy as np
import matplotlib.pyplot as plt

k_opt = 1.248
d_H = np.log(2) / np.log(2.502907875)
t_P = 5.391e-44
t_now = 4.354e17
n_0 = t_now / t_P

print(f"d_H = ln(2)/ln(α_F) = {d_H:.6f}")
print(f"n_0 = t_now/t_P = {n_0:.4e}")
print(f"ln(n_0) = {np.log(n_0):.4f}")
```

## 3.1 Verify the key identity: $(1/\varepsilon)(d\varepsilon/dn) = -2/(n\ln n)$

```python
n_test = np.logspace(3, 18, 500)
eps = k_opt / np.log(n_test + 1)**2

# Numerical derivative
d_eps = np.gradient(eps, n_test)
frac_rate_numerical = d_eps / eps

# Analytic prediction
frac_rate_analytic = -2 / (n_test * np.log(n_test + 1))

# Compare
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.loglog(n_test[10:-10], np.abs(frac_rate_numerical[10:-10]), 'b-', lw=1.5, label='Numerical')
plt.loglog(n_test, np.abs(frac_rate_analytic), 'r--', lw=1, label=r'$-2/(n\ln n)$')
plt.xlabel('n'); plt.ylabel(r'$|(1/\varepsilon)(d\varepsilon/dn)|$')
plt.title('Key identity verification'); plt.legend(); plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
ratio = frac_rate_numerical[20:-20] / frac_rate_analytic[20:-20]
plt.semilogx(n_test[20:-20], ratio, 'g-', lw=1)
plt.axhline(1.0, color='r', ls='--')
plt.xlabel('n'); plt.ylabel('Ratio (numerical/analytic)')
plt.title('Accuracy'); plt.ylim(0.99, 1.01); plt.grid(alpha=0.3)

plt.tight_layout(); plt.show()

print(f"Mean ratio = {np.mean(ratio):.8f} (should be ~1.000)")
```

## 3.2 Full Hubble evolution over 60 decades

```python
# Calibrated from data
H_inf = 104.5    # km/s/Mpc
beta = -624304.0

n_cosmic = np.logspace(3, 61, 500)
xi = 1 / np.log(n_cosmic)**2
H = H_inf + beta * xi

plt.figure(figsize=(10, 5))
plt.semilogx(n_cosmic, H, 'b-', lw=2)
plt.axhline(H_inf, color='gray', ls=':', label=f'$H_\\infty = {H_inf}$')
plt.axhline(72.7, color='green', ls='--', label='$H_0 \\approx 72.7$')
plt.axhline(67.4, color='orange', ls='--', label='Planck = 67.4')
plt.axvline(n_0, color='gray', ls=':', alpha=0.5)
plt.text(n_0*2, 80, 'today', fontsize=9, color='gray')
plt.xlabel('Planck tick n'); plt.ylabel('H [km/s/Mpc]')
plt.title(r'Hubble evolution: $H(n) = H_\infty + \beta/\ln^2(n)$')
plt.legend(); plt.ylim(-200, 200); plt.grid(alpha=0.3)
plt.show()
```

## Conclusion

The identity $(1/\varepsilon)(d\varepsilon/dn) = -2/(n\ln n)$ is verified to 8+ digits. This is the **key step** connecting symplectic dynamics to Hubble evolution.
