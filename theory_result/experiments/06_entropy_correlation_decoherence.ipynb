# Notebook 6: Entropy, Correlation Length, and Decoherence

**Paper reference**: §II.I-K

**Goal**: Compute the three derived physical quantities from the symplectic lattice framework.

```python
import numpy as np
import matplotlib.pyplot as plt

k_opt = 1.248
A = np.log(2)  # Lyapunov prefactor
t_P = 5.391e-44
l_P = 1.616e-35
c = 2.998e8
hbar = 1.055e-34
kB = 1.381e-23
t_now = 4.354e17
n_0 = t_now / t_P
ln_n0 = np.log(n_0)
```

## 6.1 KS Entropy Rate: $h_{KS}(n) = A\sqrt{k}/\ln(n)$

```python
n_range = np.logspace(2, 61, 500)
h_KS = A * np.sqrt(k_opt) / np.log(n_range)

plt.figure(figsize=(10, 4))
plt.loglog(n_range, h_KS, 'b-', lw=2)
plt.axvline(n_0, color='gray', ls=':', alpha=0.5)
plt.xlabel('Planck tick n'); plt.ylabel(r'$h_{KS}$ [nats/tick]')
plt.title(r'Kolmogorov-Sinai entropy rate: $h_{KS} = A\sqrt{k}/\ln(n)$')
plt.grid(alpha=0.3)
plt.show()

h_now = A * np.sqrt(k_opt) / ln_n0
S_total = A * np.sqrt(k_opt) * n_0 / ln_n0
print(f"h_KS(now) = {h_now:.6e} nats/tick")
print(f"S_total = {S_total:.4e} nats = {S_total/np.log(2):.4e} bits")
print(f"\nInterpretation: the Universe produces information ever more slowly,")
print(f"asymptotically freezing into a zero-entropy state — dynamical heat death.")
```

## 6.2 Correlation Length: $\xi_{\rm corr} = \ln^2(n)/k$ lattice spacings

```python
xi_corr = np.log(n_range)**2 / k_opt * l_P  # in meters
R_hubble = c * n_range * t_P  # Hubble radius in meters

plt.figure(figsize=(10, 5))
plt.loglog(n_range, xi_corr, 'b-', lw=2, label=r'$\xi_{\rm corr} = \ln^2(n)/k \cdot l_P$')
plt.loglog(n_range, R_hubble, 'r--', lw=1.5, label=r'$R_H = c \cdot n \cdot t_P$')
plt.fill_between(n_range, xi_corr, R_hubble, alpha=0.05, color='red')
plt.axvline(n_0, color='gray', ls=':', alpha=0.5)
plt.xlabel('Planck tick n'); plt.ylabel('Length [m]')
plt.title(r'Correlation length vs Hubble radius: $\xi \ll R_H$ always')
plt.legend(); plt.grid(alpha=0.3)
plt.show()

xi_now = ln_n0**2 / k_opt * l_P
R_H_now = c * t_now
print(f"ξ_corr(now) = {xi_now:.4e} m ({ln_n0**2/k_opt:.0f} Planck lengths)")
print(f"R_H(now)    = {R_H_now:.4e} m")
print(f"Ratio ξ/R_H = {xi_now/R_H_now:.4e}")
print(f"\nImplication: lattice dynamics are LOCAL — no horizon problem.")
print(f"CMB homogeneity follows from universality of the Feigenbaum cascade.")
```

## 6.3 Decoherence Time: $\tau_{\rm dec} = \ln(n)/(A\sqrt{k}) \cdot t_P$

```python
tau_dec = np.log(n_range) / (A * np.sqrt(k_opt)) * t_P

plt.figure(figsize=(10, 4))
plt.loglog(n_range, tau_dec, 'b-', lw=2)
plt.loglog(n_range, np.full_like(n_range, t_P), 'r--', label=r'$t_P$')
plt.axvline(n_0, color='gray', ls=':', alpha=0.5)
plt.xlabel('Planck tick n'); plt.ylabel(r'$\tau_{\rm dec}$ [s]')
plt.title(r'Quantum decoherence time: $\tau = \ln(n)/(A\sqrt{k}) \cdot t_P$')
plt.legend(); plt.grid(alpha=0.3)
plt.show()

tau_now = ln_n0 / (A * np.sqrt(k_opt)) * t_P
print(f"τ_dec(now) = {tau_now:.4e} s = {tau_now/t_P:.1f} × t_P")
print(f"\nThis is ~140 Planck times — far shorter than any lab timescale,")
print(f"explaining why gravity appears classical at macroscopic scales.")
print(f"\nFalsifiable: if massive-particle interferometry measures τ_dec < 100 t_P,")
print(f"the cooling law is ruled out.")
```
